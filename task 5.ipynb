```python
# ==============================
# Program 1: Social Media – Post Engagement Analyzer
# ==============================

def analyze_post_engagement(likes_data):
    total_likes = sum(likes_data)

    if total_likes >= 1000:
        engagement_status = "Viral Post"
    else:
        engagement_status = "Normal Engagement"

    print("\n--- Post Engagement Report ---")
    print(f"Total Likes Received: {total_likes}")
    print(f"Engagement Status: {engagement_status}")


```

```python
# ==============================
# Program 2: Healthcare – Medicine Stock Alert System
# ==============================

def check_medicine_stock_level(current_stock):
    if current_stock < 10:
        stock_status = "Low Stock Alert"
    else:
        stock_status = "Stock Sufficient"

    print("\n--- Medicine Stock Report ---")
    print(f"Available Medicine Stock: {current_stock}")
    print(f"Stock Status: {stock_status}")



```

```python
# ==============================
# Program 3: Agriculture – Rainfall Adequacy Checker
# ==============================

def evaluate_rainfall(rainfall_readings, required_average):
    total_rainfall = sum(rainfall_readings)
    average_rainfall = total_rainfall / len(rainfall_readings)

    if average_rainfall >= required_average:
        rainfall_status = "Adequate Rainfall"
    else:
        rainfall_status = "Inadequate Rainfall"

    print("\n--- Rainfall Analysis Report ---")
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {rainfall_status}")




```

```python
# ==============================
# Program 4: Social Media – Duplicate Account Detection
# ==============================

def check_duplicate_accounts(user_list):
    unique_usernames = set(user_list)

    if len(unique_usernames) != len(user_list):
        duplicate_found = "Yes"
    else:
        duplicate_found = "No"

    print("\n--- Duplicate Account Check ---")
    print(f"Duplicate Accounts Found: {duplicate_found}")


```

```python
# ==============================
# Program 5: Healthcare – Appointment Eligibility Checker
# ==============================

def check_patient_eligibility(age):
    if age >= 18:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"

    print("\n--- Appointment Eligibility Report ---")
    print(f"Patient Age: {age}")
    print(f"Appointment Eligibility: {eligibility_status}")


```

```python
# ==============================
# Program 6: Agriculture – Premium Crop Price Filter
# ==============================

def get_premium_crops(price_list):
    premium_crops = []

    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    print("\n--- Premium Crop Filter ---")
    print(f"Premium Crop Prices: {premium_crops}")


```

```python
# ==============================
# Program 7: System Monitoring – Application Health Checker
# ==============================

def monitor_system_health(error_count):
    if error_count == 0:
        system_status = "Healthy"
    elif error_count <= 5:
        system_status = "Minor Issues"
    else:
        system_status = "Critical Issues"

    print("\n--- System Health Report ---")
    print(f"Total Errors Detected: {error_count}")
    print(f"System Health Status: {system_status}")


```

```python
# ==============================
# Program 8: Banking – Daily Transaction Limit Checker
# ==============================

def verify_transaction(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        transaction_status = "Approved"
    else:
        transaction_status = "Rejected"

    print("\n--- Transaction Verification ---")
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {transaction_status}")


```

```python
# ==============================
# Program 9: E-Learning – Student Attendance Eligibility System
# ==============================

def check_exam_eligibility(attendance_records):
    total_days = len(attendance_records)
    present_days = attendance_records.count("P")

    attendance_percentage = (present_days / total_days) * 100

    if attendance_percentage >= 75:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"

    print("\n--- Attendance Report ---")
    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility Status: {eligibility_status}")


```

```python
# ==============================
# Program 10: Smart Electricity Bill Analyzer
# ==============================

def calculate_bill(units_consumed):
    if units_consumed <= 100:
        total_bill = units_consumed * 3
    elif units_consumed <= 200:
        total_bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        total_bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

    if total_bill < 500:
        usage_category = "Low Usage"
    elif total_bill <= 1500:
        usage_category = "Moderate Usage"
    else:
        usage_category = "High Usage"

    print("\n--- Electricity Bill Report ---")
    print(f"Total Electricity Bill: {total_bill}")
    print(f"Usage Category: {usage_category}")
```
